# Simulation Mag

```python
mmag = mtrue.copy()
mmag[mtrue==1] = 0.05
mmag[mtrue==2] = 0.15
```

```python
simulation_mag = PF.magnetics.simulation.Simulation3DIntegral(
    survey=survey_mag,
    mesh=mesh,
    chiMap=maps.IdentityMap(nP=mesh.nC),
    actInd=np.ones_like(mtrue,dtype='bool'),
    #store_sensitivities="forward_only",
)
noise_floor = 1.
relative_error = 0.
# Compute predicted data for some model
dpred_mag = simulation_mag.make_synthetic_data(mmag,relative_error=relative_error,noise_floor=noise_floor, f=None, add_noise=True)
```

```python
fig = plt.figure(figsize=(10,10))

#mesh.plotSlice(np.ones(mesh.nC),normal='Z', ind=-1, grid=True,pcolorOpts={'cmap':'Greys'},ax=plt.gca())
plt.gca().set_aspect('equal')
plt.gca().scatter(X,Y,s=1)
rxLoc = np.c_[utils.mkvc(X.T), utils.mkvc(Y.T), utils.mkvc(Z.T)]
mm = utils.plot2Ddata(rxLoc,dpred_mag.dobs,
                      ax=plt.gca(),level=True, 
                     nx=50,ny=50, dataloc=True,ncontour=100, shade=True,
                      contourOpts={'cmap':'viridis'},
                      levelOpts={'colors':'k','linestyles':'dotted','linewidths':0.25,}
                      )
plt.colorbar(mm[0])
plt.gca().set_title('MAGNETIC RESPONSE')
```

# Simulation Remanence

```python
# inclination, declination of remanent block
M_inc_dec = [-10.0, 45.0]
M_xyz = 0.1 * utils.mat_utils.dip_azimuth2cartesian(*M_inc_dec)

print(M_xyz)
# Assign magnetization values

m_rem = np.zeros((len(mesh), 3), order='F')
m_rem[mtrue==1] = M_xyz
m_rem = m_rem.reshape(-1, order='F')
```

```python
# Create the simulation
simulation_rem = PF.magnetics.simulation.Simulation3DIntegral(
    survey=survey_mag,
    mesh=mesh, 
    chiMap=maps.IdentityMap(nP=3*len(mesh)),
    actInd=np.ones_like(mtrue,dtype='bool'),
    modelType="vector",
    store_sensitivities="forward_only"
)

dpred_rem = simulation_rem.make_synthetic_data(
    m_rem,
    noise_floor=1.0,
    add_noise=True
)
```

```python
fig = plt.figure(figsize=(10,10))
data_total_mag = dpred_rem.dobs + dpred_mag.dobs

#mesh.plotSlice(np.ones(mesh.nC),normal='Z', ind=-1, grid=True,pcolorOpts={'cmap':'Greys'},ax=plt.gca())
plt.gca().set_aspect('equal')
plt.gca().scatter(X,Y,s=1)
rxLoc = np.c_[utils.mkvc(X.T), utils.mkvc(Y.T), utils.mkvc(Z.T)]
mm = utils.plot2Ddata(rxLoc,data_total_mag,
                      ax=plt.gca(),level=True, 
                      nx=50,ny=50, dataloc=True,ncontour=100, shade=True,
                      contourOpts={'cmap':'viridis'},
                      levelOpts={'colors':'k','linestyles':'dotted','linewidths':0.25,}
                      )
plt.colorbar(mm[0])
plt.gca().set_title('MAGNETIC RESPONSE')
```

```python
plt.figure(figsize=(10,4))
ax = plt.subplot(121)
mesh.plot_slice(
    m_rem.reshape((-1, 3), order='F'),
    v_type='CCv',
    view='vec',
    normal='y',
    ax=ax,
    quiver_opts = {'scale':2.0}
)
ax.set_xlim([-2500, 2500])
ax.set_ylim([-2000 , 0])

ax = plt.subplot(122)
mesh.plot_slice(
    m_rem.reshape((-1, 3), order='F'),
    v_type='CCv',
    view='vec',
    normal='z',
    ind = -5,
    ax=ax,
    quiver_opts = {'scale':2.0}
)
ax.set_xlim([-2500, 2500])
ax.set_ylim([-2500, 2500])
```

# Simulation Grav

```python
# Background of 2.9 g/cc
mden = mtrue.copy()
mden[mtrue==1] = 0.3
mden[mtrue==2] = -0.2
```

```python
simulation_grav = PF.gravity.simulation.Simulation3DIntegral(
    survey=survey_gravity,
    mesh=mesh,
    rhoMap=maps.IdentityMap(nP=mesh.nC),
    actInd=np.ones_like(mtrue,dtype='bool'),
)
noise_floor = 0.01
relative_error = 0.
# Compute predicted data for some model
dpred_grav = simulation_grav.make_synthetic_data(mden,relative_error=relative_error,noise_floor=noise_floor, f=None, add_noise=True)
```

```python
fig = plt.figure(figsize=(10,10))

#mesh.plotSlice(np.ones(mesh.nC),normal='Z', ind=-1, grid=True,pcolorOpts={'cmap':'Greys'},ax=plt.gca())
plt.gca().set_aspect('equal')
plt.gca().scatter(X,Y,s=1)
rxLoc = np.c_[utils.mkvc(X.T), utils.mkvc(Y.T), utils.mkvc(Z.T)]
mm = utils.plot2Ddata(rxLoc,-dpred_grav.dobs,
                      ax=plt.gca(),level=True, 
                     nx=50,ny=50, dataloc=True,ncontour=100, shade=True,
                      contourOpts={'cmap':'viridis'},
                      levelOpts={'colors':'k','linestyles':'dotted','linewidths':0.25}
                      )
plt.colorbar(mm[0])
plt.gca().set_title('GRAVITY RESPONSE')
```
