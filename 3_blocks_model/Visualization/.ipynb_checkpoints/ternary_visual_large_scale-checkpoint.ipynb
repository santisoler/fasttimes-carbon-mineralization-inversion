```python
from scipy.stats import kde

# Create data: 200 points
data = np.c_[density,np.log(magsus+1e-7)]
indx = 1-np.isnan(data).sum(axis=1)
data = data[np.where(indx==1)[0],:]
x = data[:,0]
y = data[:,1]
 
# Create a figure with 6 plot areas
fig, axes = plt.subplots(ncols=6, nrows=1, figsize=(21, 5))
 
# Everything sarts with a Scatterplot
axes[0].set_title('Scatterplot')
axes[0].plot(x, y, 'ko')
# As you can see there is a lot of overplottin here!
 
# Thus we can cut the plotting window in several hexbins
nbins = 20
axes[1].set_title('Hexbin')
axes[1].hexbin(x, y, gridsize=nbins, cmap=plt.cm.BuGn_r)
 
# 2D Histogram
axes[2].set_title('2D Histogram')
axes[2].hist2d(x, y, bins=nbins, cmap=plt.cm.BuGn_r)
 
# Evaluate a gaussian kde on a regular grid of nbins x nbins over data extents
k = kde.gaussian_kde(data[::2,:2].T)
xi, yi = np.mgrid[x.min():x.max():nbins*1j, y.min():y.max():nbins*1j]
zi = k(np.vstack([xi.flatten(), yi.flatten()]))
 
# plot a density
#axes[3].set_title('Calculate Gaussian KDE')
axes[3].pcolormesh(xi, yi, zi.reshape(xi.shape), cmap=plt.cm.BuGn_r)
 
# add shading
axes[4].set_title('2D Density with shading')
axes[4].pcolormesh(xi, yi, zi.reshape(xi.shape), shading='gouraud', cmap=plt.cm.BuGn_r)
 
# contour
axes[5].set_title('Contour')
axes[5].pcolormesh(xi, yi, zi.reshape(xi.shape), shading='gouraud', cmap=plt.cm.BuGn_r)
axes[5].contour(xi, yi, zi.reshape(xi.shape) )
```

```python
class ColorMap2D:
    """
    Many thanks to:
    - xkcd for their colors palette (https://xkcd.com/color/rgb/)
    - Minute Physics for their explanation on how to compute proper color mix (https://www.youtube.com/watch?v=LKnqECcg6Gw)
    """
    background_color = np.r_[1.0, 1.0, 0.8941176470588236, 1.] #off white xkcd
    density_color = np.array([0.00390625, 0.39453125, 0.984375, 1.  ]) #bright blue xkcd
    magnetic_color = np.array([0.984375  , 0.58984375, 0., 1.        ]) #orange -> bright blue complementary color
    mix_color = np.r_[0,1,0] #green only available with barycenter
    interpolation = 'sigmoid' #linear, sigmoid
    use_barycenter = False #in progress
    sigmoid_tol = 1e-3
        
    def __init__(self,mesh ,denmodel, magmodel, actvMap,maxden=None,maxmag=None, **kwargs):
        
        self.denmodel = denmodel-denmodel.min()
        self.denmin = denmodel.min()
        self.magmodel = magmodel
        self.actvMap = actvMap
        self.mesh = mesh
        self.maxden = maxden-denmodel.min()
        self.maxmag = maxmag
        Sutils.setKwargs(self, **kwargs)
    
    def sigmoid(self, x, x0, k):
        y = 1 / (1 + np.exp(-k*(x-x0)))
        return y
    
    def barycenter(self,denvalue,magvalue):
        center = center = np.c_[
            [0,0],
            [1,0],
            [0,1],
            [1,1]
        ].T
        
        weight0 = self.interpolate(denvalue,maxden)
        weight1 = self.interpolate(magvalue,maxmag)
        dist, _ = Sutils.pgi_utils.ComputeDistances(np.c_[weight0,weight1],center)
        dist = 1-dist
        return dist
    
    def interpolate(self, value, maxval=None):
        
        if maxval is None:
            maxval = np.abs(value).max()
        
        if self.interpolation == 'linear':
            weight = np.minimum(1, np.abs(value/(maxval)))
        else:
            from scipy.optimize import curve_fit
            param0, _ =curve_fit(self.sigmoid,
            np.r_[
                -2*(np.abs(value).max()),
                np.abs(value).min(),maxval,
                2*(maxval),], #Density
            np.r_[0.,self.sigmoid_tol,1-self.sigmoid_tol,1.,], #color
            )
            f0 = lambda x: np.sqrt(
            sigmoid(x,*param0)
            )
            weight = self.sigmoid(np.abs(value),*param0)
        
        return weight
    

    def fden(self,denvalue):
        weight = self.interpolate(denvalue,self.maxden)
        return np.sqrt((weight)*(self.density_color[:,np.newaxis]**2) + ((1-weight)*(self.background_color[:,np.newaxis]**2))).T
    
    
    def fmag(self,magvalue):
        
        weight = self.interpolate(magvalue,self.maxmag)           
        return np.sqrt((weight)*(self.magnetic_color[:,np.newaxis]**2) + ((1-weight)*(self.background_color[:,np.newaxis]**2))).T
    
    
    def fcolor(self,denvalue,magvalue):
        
        if self.use_barycenter:
            weights = self.barycenter(denvalue,magvalue)
            color = np.sqrt((1./(np.sum(weights,axis=1))[:,np.newaxis])*(weights[:,0][:,np.newaxis]*(self.background_color**2) + weights[:,1][:,np.newaxis]*(self.fden(denvalue)**2) + weights[:,2][:,np.newaxis]*(self.fmag(magvalue)**2) + weights[:,3][:,np.newaxis]*(self.mix_color**2))) 
        else:
            weight0 = self.interpolate(denvalue,self.maxden)[:,np.newaxis]
            weight1 = self.interpolate(magvalue,self.maxmag)[:,np.newaxis]
            weight2 = np.ones_like(weight1)*np.maximum(1e-6,(weight0+weight1))
            color =  np.sqrt((1/(weight2))*(weight0*(self.fden(denvalue)**2) + weight1*(self.fmag(magvalue)**2) + 1e-6*(self.fmag(magvalue)**2)))
        return np.minimum(color,1)
    
    def plot_colormap(self,ax=None, opts={}):
        
        # Plot points
        a, b = np.mgrid[self.denmodel.min():self.denmodel.max():0.01, 
                   self.magmodel.min():self.magmodel.max():1e-4, 
        ]
        if ax is None:
            fig,ax = plt.subplots(1,1,figsize=(5,5))
            
        ax.scatter(a.flatten(),b.flatten(),c=self.fcolor(a.flatten(),b.flatten()), **opts)
        ax.set_xlim([a.min(),a.max()])
        ax.set_ylim([b.min(),b.max()])
        
    def plot_scatter(self, ax=None, opts={'s':1,'edgecolors':'k','linewidths':0.1}):
        if ax is None:
            fig,ax = plt.subplots(1,1,figsize=(5,5))
        
        ax.scatter(
            self.denmodel[self.actvMap],self.magmodel[self.actvMap],
            c=color2dmap.fcolor(self.denmodel[self.actvMap],self.magmodel[self.actvMap]),
            **opts
        )
        ax.set_xticks(np.round(np.linspace(0,self.denmodel[self.actvMap].max(),7),2))
        ax.set_xticklabels(np.round(np.round(np.linspace(0,self.denmodel[self.actvMap].max(),7),2)+self.denmin,2))
        
    def plot_model(self, ax=None):
        ternarymap = self.fcolor(self.denmodel,self.magmodel)
        ternaryimage = np.vstack(
            [(self.actvMap*ternarymap[:,0]),
            (self.actvMap*ternarymap[:,1]),
            (self.actvMap*ternarymap[:,2]),
            (self.actvMap*ternarymap[:,3]),]
        ).T
        #fix artifact
        ternaryimage[np.where(np.all(ternaryimage==1,axis=1))] = self.background_color
        ternaryimage[np.where(np.all(np.isnan(ternaryimage),axis=1))] = np.r_[1,1,1,0]
        if ax is None:
            fig,ax = plt.subplots(1,1,figsize=(10,10))
            # Turns off grid on the left Axis.
        ax.grid(False)

        #plt.gca().axis('off')
        ax.set_aspect(1)
        
        ax.imshow(((ternaryimage).reshape(self.mesh.nCx,self.mesh.nCy,4,order='F')[:,::-1,:]).transpose(1,0,2),
                  extent = [self.mesh.vectorCCx.min(),self.mesh.vectorCCx.max(),self.mesh.vectorCCy.min(),self.mesh.vectorCCy.max()], #(left, right, bottom, top)
                  zorder=0
            )
```
